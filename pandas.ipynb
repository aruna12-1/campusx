Fill Missing Values for Float (Numerical Columns)

Usually we fill with:

✅ Mean
Fill with Mean Used when data is normally distributed

✅ Median
Fill with Median (Better if outliers present)

✅ 0 (sometimes)

2️⃣ Fill Missing Values for String (Categorical Columns)

Usually we fill with:

✅ Mode (most frequent value)

✅ "Unknown"

✅ "Missing"

Numerical → Fill with mean / median

Categorical → Fill with mode
